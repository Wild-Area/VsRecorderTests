```julia
import Pkg
Pkg.activate("..")

using VsRecorderBase, VsRecorder

VsRecorder.download_all_ocr_languages()
```

```julia
frame = VsFrame(image = VsRecorder.load_data("scenes", "zhs", "team-preview-selecting.png"))
```

```julia
using BenchmarkTools
```

```julia
ctx = VsRecorder.default_context();
```

```julia
@btime vs_tryparse_scene(VsRecorder.Scenes.TeamPreviewSelecting, $frame, $ctx, force = true)
```

```julia
frame_gray = VsFrame(image = Gray.(VsRecorder.load_data("scenes", "zhs", "team-preview-selecting.png")))
```

```julia
ctx_gray = VsRecorderBase.initialize(vs_setup(
    VsRecorder.PokemonBattle, parse_type = VsRecorder.PARSE_BOTH_PLAYERS, set_global = false,
    init_descriptors = false, use_gray_image = true
));
```

```julia
@btime vs_tryparse_scene(VsRecorder.Scenes.TeamPreviewSelecting, $frame_gray, $ctx_gray, force = true)
```

```julia
frame2 = VsFrame(image = VsRecorder.load_data("scenes", "zhs", "team-preview.png"))
```

```julia
@btime vs_tryparse_scene(VsRecorder.Scenes.TeamPreview, $frame2, $ctx, force = true)
```
