```julia
using VsRecorderBase, VsRecorder
using VsRecorder.Scenes
using BenchmarkTools
```

```julia
img = load_image(VsRecorder.datapath("scenes", "zhs", "team-view.png"))
frame = VsFrame(image = img)
```

```julia
setup = vs_setup(PokemonBattle, language = VsI18n.ZHS, init_descriptors = false)
ctx = VsRecorderBase.initialize(setup);
```

```julia
result = @btime vs_tryparse_scene(Scenes.TeamView, $frame, $ctx, force = true);
```

```julia
team = result.team
@show team.title
@show team.author
@show team.rental_code
exported = export_team(team)
println(exported)
```
